```
alpha = 0.01
slope = 2
intercept = 1
m = 20
x = 30*np.random.random((m,1))+(-15)
print(x.shape)
x_input = np.append(np.ones((m,1)),x,axis=1)
print(x_input)
```

```
y = slope*x+intercept
print(y)
```

```
def calculate_loss(h,y):
  k = (1/(2*m))*np.sum(np.power((h-y),2))
  return k
```
